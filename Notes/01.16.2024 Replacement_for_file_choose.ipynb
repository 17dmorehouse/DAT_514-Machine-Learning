# Replacement for file.choose() using Tkinter

I hate the process of typing the file name with the full file path into a piece of code,   
I always make typing errors.  Then there is the difference in the number of slashes used between
Windows and Mac

In R,  I habitually use file.choose() to get the file name and path.

There is no file.choose() in Python,  but there is the tkinter window utility available.

You can program all kinds of windowing options for GUIs using tkinter that run in Windows or Mac OS, or Linux.

Here is the code for a file chooser in Python

```python
import tkinter as Tk # from tkinter import Tk for Python 3.x
from tkinter.filedialog import askopenfilename
from tkinter import filedialog

root = Tk.Tk()
root.withdraw()

file_path = filedialog.askopenfilename()
file_path
```
